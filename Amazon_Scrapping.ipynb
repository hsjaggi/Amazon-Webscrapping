```python
#from urllib.parse import urlparse
import os
import time
#import requests
from selenium import webdriver
import time
import math, requests, json
from datetime import datetime
from datetime import date

from gspread_pandas import Spread,Client
from datetime import timedelta
#from sqlalchemy import create_engine
#import psycopg2 
import io
import numpy as np
import pandas as pd

#import warnings
#warnings.filterwarnings('ignore')
import re
# import schedule
from bs4 import BeautifulSoup

# import schedule
import time
import json


now = datetime.now()
d_1=date.today()

d_11=d_1.strftime("%Y-%m-%d")
print(d_11)

chrome_options = webdriver.ChromeOptions()
chrome_options.binary_location = os.environ.get("GOOGLE_CHROME_BIN")
chrome_options.add_argument("--headless")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--no-sandbox")

browser = webdriver.Chrome()

hour=int(datetime.now().strftime('%H'))

url = "https://www.amazon.in/dp/B0BR58HRD2"
browser.get(url)
time.sleep(2)

title=[]
price=[]
ASIN=[]
rating=[]
discount=[]

asins=['B0BVBWHL2F','B0BRKXFVVJ','B0BRKXBRMF','B0BRJ5QH8G','B09YD879Y4','B0C3MHB63Q']

for i in asins:
    url = "https://www.amazon.in/dp/" + str(i)
    browser.get(url)
    time.sleep(3)
    
    html=browser.page_source
    soup = BeautifulSoup(html, "html.parser")

    #--------------------------#
    try:
        title.append(soup.find(id="productTitle").get_text(strip=True))
    except:
        title.append('None')
    try:
        price.append(soup.find("span", {"class": "a-price-whole"}).get_text())
    except:
        price.append('None')
#     #---------------------------#        
    try:
        rating.append(soup.find("span", {"class": "a-icon-alt"}).get_text().split('out of')[0].strip())
    except:
        rating.append('None')    
    #---------------------------#        
    try:
        ASIN.append(i)
    except:
        ASIN.append('None')
    try:
        discount.append(soup.find('span', attrs={"class": "a-size-large a-color-price savingPriceOverride aok-align-center reinventPriceSavingsPercentageMargin savingsPercentage", "aria-hidden": "false"}).get_text())
    except:
        discount.append('None')

df=pd.DataFrame(list(zip(title,price,rating,ASIN,discount)),columns=['Title','Price','Rating','ASIN','Discount'])

df['Date'] = d_11

df = df[['Date','Title', 'Price', 'Rating', 'ASIN', 'Discount']]

path = r"C:\\Users\\91941\\myprojects\\Jupyter_notebooks\\Projects\\Untitled Folder\\durable-verve-328507-160feca63418.json"

with open(f'{path}', 'r') as f:
    data = json.load(f)

spread = Spread('Amazon_Data1',config=data) ## worksheet_name
sheet1=spread.open_sheet("Consumer_durables") ##sheet_name
# table_updated=spread.sheet_to_df(sheet=sheet1,index=0)
# x=table_updated.shape[0]

# spread.df_to_sheet(df,index=False,start=(1,1),headers=True)
spread.df_to_sheet(df,index=False,start=(1,1),headers=True)
print('Pushed to ghseet')
print(d_11)
```
